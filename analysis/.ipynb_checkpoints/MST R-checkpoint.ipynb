```r
library(tidyverse)
library(MASS)
library(GSAR)
library(parallel)
```

```r
path  <- "/home/jovyan/work/mhar_thesis_sims_points/"
save_path  <- "/home/jovyan/work/mhar_zvalues/"

simplex_gen  <- function(n, samples){
    simplex_out  <- matrix(, nrow = samples, ncol = n)
    for(r in c(1:samples)){
        sample  <- runif(n-1,0,1)  %>%  
                    sort(decreasing=FALSE)
        sample  <- append(0.0, sample)  %>% 
                    append(1.0)
        sample  <- sample[-1] - sample
        sample  <- sample[1:n]

        simplex_out[r,]  <- sample
     }
    return(simplex_out)
}

mst  <- function(file_, nsamp=10000, perm=50){
    # Read files with samples
    df_  <- read_csv(file_)  %>%  
    dplyr::select(starts_with('x_'))  %>% 
    sample_n(size=nsamp, replace=FALSE)
    # Get dim of the read file
    n_rows  <- dim(df_)[1]
    n_cols  <- dim(df_)[2]
        if (n_rows > (nsamp-1)){
        simplex_sample  <- suppressWarnings(simplex_gen(n_cols,n_rows))
        
        df_  <- df_  %>% 
                data.matrix()   %>% 
                unname()
        
        df_  <- rbind(df_, simplex_sample)  
        df_  <- df_ %>% 
                aperm(c(2,1)) 
        result <- WWtest(object=df_, group=c(rep(1, n_rows),rep(2, n_rows)),
                         nperm=perm, pvalue.only=FALSE)
        z_df  <- tibble(dim=n_cols, z=result$statistic, 
                        p_value=result$p.value,
                             nsamp=nsamp)
        gc()
        return(z_df)
        
    }
    
}


# For dim 5
vv  <- c(6:7)
set.seed(15523)
files <- list.files(path, pattern='2_5_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, mst, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_5_ee.csv", sep=''))
gc()


# For dim 15
set.seed(12153)
files <- list.files(path, pattern='2_15_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, mst, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_15_ee.csv", sep=''))
gc()

# For dim 25
set.seed(12325)
files <- list.files(path, pattern='2_25_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, mst, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_25_ee.csv", sep=''))
gc()

# For dim 50
set.seed(1233)
files <- list.files(path, pattern='2_50_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, mst, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_50_ee.csv", sep=''))
gc()




#########################################
####################################################
######################################################

path  <- "/home/jovyan/work/mhar_thesis_sims_points/"
save_path  <- "/home/jovyan/work/mhar_zvalues/"


hyper_gen  <- function(n, samples){
    simplex_out  <- matrix(, nrow = samples, ncol = n)
    for(r in c(1:samples)){
        sample  <- runif(n,-1,1) 
        sample  <- sample[1:n]
        simplex_out[r,]  <- sample
     }
    return(simplex_out)
}


msth  <- function(file_, nsamp=10000, perm=50){
    # Read files with samples
    df_  <- read_csv(file_)  %>%  
    dplyr::select(starts_with('x_'))  %>% 
    sample_n(size=nsamp, replace=FALSE)
    # Get dim of the read file
    n_rows  <- dim(df_)[1]
    n_cols  <- dim(df_)[2]
        if (n_rows > (nsamp-1)){
        simplex_sample  <- suppressWarnings(hyper_gen(n_cols,n_rows))
        
        df_  <- df_  %>% 
                data.matrix()   %>% 
                unname()
        
        df_  <- rbind(df_, simplex_sample)  
        df_  <- df_ %>% 
                aperm(c(2,1)) 
        result <- WWtest(object=df_, group=c(rep(1, n_rows),rep(2, n_rows)),
                         nperm=perm, pvalue.only=FALSE)
        z_df  <- tibble(dim=n_cols, z=result$statistic, 
                        p_value=result$p.value,
                             nsamp=nsamp)
        gc()
        return(z_df)
        
    }
    
}


vv  <- c(6:7)
# For dim 5
set.seed(15523)
files <- list.files(path, pattern='1_5_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, msth, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_5_e_hyper.csv", sep=''))
gc()


# For dim 15
set.seed(12153)
files <- list.files(path, pattern='1_15_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, msth, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_15_e_hyper.csv", sep=''))
gc()

# For dim 25
set.seed(12325)
files <- list.files(path, pattern='1_25_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, msth, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_25_e_hyper.csv", sep=''))
gc()

# For dim 50
set.seed(1233)
files <- list.files(path, pattern='1_50_.*')
files  <- paste(path, files, sep='')
files  <- files[vv]
#mst(path, files[1], 5,2)
results <- mclapply(files, msth, mc.cores = 2)
results  <- bind_rows(results)
results  %>% write_csv(paste(save_path,"z_values_50_e_hyper.csv", sep=''))
gc()
```
